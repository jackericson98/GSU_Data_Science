    class which contains a node class.Just the way we explored it in the Lecture, your
    node class will house the data (integer in this case) and a pointer to the next

    Populate your linked list with the following integers and print it .
    
    50, 11, 33, 21, 40, 71
    
    No you do not need to print the commas ;)
    
    Delete N-th node from the end of the linked list and print the linked list after deletion. Here N = 2
    
    Below is the expected output after deleting the second last element.
    
    50, 11, 33, 21, 71
    
    ATTN : Note : Here we do not know the length of the list.
    Complete the above deletion operation without calculating the length of the list.
    Your solution should only make a single pass through the linked list, adhering
    to O(n) time complexity overall and O(1) space complexity.

```python
# Node class used to initialize a node. Stores the data for the node and the next node in the list.
class Node:
    
    # Initialize node class. 
    def __init__(self, data):
        # Data input
        self.data = data
        # Defaults to None for node.next
        self.next = None
        
# Linked List class with methods that create list, add nodes, print list and delete the nth element from the end.       
class LinkedList:
    
    # Initialize Linked List
    def __init__(self):
        
        # Defaults to empty list. The head pointer is set to None.
        self.head = None

    # addNode Method to add a node to the end of the list. Traverses the list until the end and once there adds our Node.
    def addNode(self, data):
        
        # Create a node with our data. 
        myNode = Node(data)
        
        # If our list is empty then set our node to the head pointer.
        if self.head is None:
            self.head = myNode
            
            # Exit the method
            return
        
        # If our list is not empty initialize a temporary variable to help us traverse the list.
        temp = self.head
        
        # Traverse the list until we hit our last element in the list.
        while temp.next:
            temp = temp.next
            
        # Set our new end node to our previous end node's next and our new end node's next equal to None
        temp.next = myNode
        myNode.next = None
        
    # Method used to add multiple nodes at once. Data must be in the order we want our list to be in.
    def addNodes(self, array):
        
        # Add each element from our data array to our linked list in order. 
        for el in array:
            self.addNode(el)
    
    # Method used to delete the nth element from the end of the list.    
    def deleteNode(self, n):
        
        # The case where user inputs a value outside the scope of what is deletable.
        if n < 1:
            print("You can't delete nothing. Please try setting n greater than 0")
            
            # Exit the method
            return
        
        
        # Initialize our fast and slow 'dummy' pointers and set them equal to the head.
        fast = self.head
        slow = self.head 
        
        
        # Move the fast pointer n spaces ahead of the slow pointer.
        for i in range(n):
            
            # Moving the pointer
            fast = fast.next
        
        # Move both pointers until our fast pointer reaches the end. 
        while fast.next:
            
            # Move pointers
            fast = fast.next
            slow = slow.next
            
        # Our slow pointer will be right behind the value we want to delete.
        
        # Set our slow pointer's next value to skip the node we want deleted.
        slow.next = slow.next.next
        
    # Method used to print each element in our list 
    def printList(self):
        
        # Hold our print variable. 
        temp = self.head
        
        # Keep printing until the next item in the list is None
        while temp:
            
            # Print node data
            print(temp.data)
            
            # Move to the next node
            temp = temp.next
            
         
# Driver Code:

# Initialize our list
myList = LinkedList()

# Add our data to the list
myList.addNodes([50, 11, 33, 21, 40, 71])

# Delete our nth to last node
myList.deleteNode(1)

# Print our entire list without the nth to last node
myList.printList()


# Uncomment below to try test cases




# # Try some test cases: Empty, single element, one element type, negative numbers, array of strings
# tests = [0, 1, 2, 3, 4, 5]
# # testing deleteing each element from our list.

# for test in tests:
#     myList = LinkedList()

#     # Add our data to the list
#     myList.addNodes([50, 11, 33, 21, 40, 71])
#     try: 
#         myList.deleteNode(test)
#         print("{} works" .format(test))
#     except:
#         print("You have an error in your code! It failed with this test case: {}" .format(test))
        
        
```
